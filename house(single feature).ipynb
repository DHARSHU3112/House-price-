```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
house=pd.read_csv("chennai_house.csv")
house
```

```python
%matplotlib inline
plt.scatter(house.Size,house.Rent,color='red') # label missing
```

```python
%matplotlib inline
plt.xlabel('Size(sqrft)',fontsize=10)
plt.ylabel('Rent(Rs)',fontsize=10)
plt.scatter(house.Size,house.Rent,color='green',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(house[['Size']],house.Rent)
```

```python
reg.predict([[400]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y=mx+b where m is slope & b is intercept m=10.27704485,x=400,b=2976.253298153033
10.27704485*400+2976.253298153033
```

```python
df=pd.read_csv("rent.csv")
df
```

```python
r=reg.predict(df)
```

```python
df['Rent']=r # creating new col in dataframe
df
```

```python
df.to_csv("Prediction_csv",index=False)
```

```python
%matplotlib inline
plt.xlabel('Size(sqrft)',fontsize=10)
plt.ylabel('Rent(Rs)',fontsize=10)
plt.scatter(house.Size,house.Rent,color='black',marker='*')
plt.plot(house.Size,reg.predict(house[['Size']]),color='blue')
```
